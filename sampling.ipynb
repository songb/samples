```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('drive/MyDrive/balance-scale.data', names=['balance', 'v1', 'v2', 'v3', 'v4'])
df.head(5)
```

```
df['balance']=df['balance']=='B'
```

```
df['balance'] = df['balance'].astype(int)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```
y=df.balance
X = df.drop('balance', axis=1)
```

```
lr = LogisticRegression().fit(X, y)
```

```
pred_y = lr.predict(X)
```

```
print(accuracy_score(pred_y, y))
```

```
pred_y
```

```
from sklearn.utils import resample
```

```
df_maj = df[df.balance==0]
df_min = df[df.balance==1]
```

```
df_min_upsample = resample(df_min, replace=True, n_samples=576, random_state=42)
```

```
df_upsample = pd.concat([df_maj, df_min_upsample])
```

```
X_upsample = df_upsample.drop('balance', axis=1)
y_upsample = df_upsample['balance']
```

```
lr_upsample = LogisticRegression().fit(X_upsample, y_upsample)
```

```
y_upsample_pred = lr_upsample.predict(X_upsample)
```

```
accuracy_score(y_upsample_pred, y_upsample)
```

```
y_upsample_pred
```

```
df_maj_downsample = resample(df_maj, replace=False, n_samples=49, random_state=42)
```

```
df_downsample = pd.concat([df_maj_downsample, df_min])
X_downsample = df_downsample.drop('balance', axis=1)
y_downsample = df_downsample['balance']
```

```
lr_downsample = LogisticRegression().fit(X_downsample, y_downsample)
```

```
y_downsample_pred = lr_downsample.predict(X_downsample)
```

```
accuracy_score(y_downsample_pred, y_downsample)
```

```
from sklearn.metrics import roc_auc_score
```

```
prob_downsample = lr_downsample.predict_proba(X_downsample)
```

```
prob_downsample_positive = [p[1] for p in prob_downsample]
```

```
roc_auc_score(y_downsample, prob_downsample_positive)
```

```
prob_upsample = lr_upsample.predict_proba(X_upsample)
prob_upsample_postitive = [p[1] for p in prob_upsample]
```

```
roc_auc_score(y_upsample, prob_upsample_postitive)
```
