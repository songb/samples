```
import pandas as pd
```

```
ids = [11, 22, 33, 44, 55, 66, 77]
countries = ['Spain', 'France', 'Spain', 'Germany', 'France']
df = pd.DataFrame(list(zip(ids, countries)), columns=['Ids', 'Countries'])
```

```
ohe = pd.get_dummies(df.Countries, prefix='Country')
```

```
df
```

```
ohe
```

```
ohe_df = pd.concat([df, ohe], axis=1)
```

```
ohe_df = ohe_df.drop('Countries', axis=1)
```
