```
import numpy as np
import pandas as pd
data = pd.read_csv("drive/MyDrive/movie_data.csv")
print(data.shape)
data.head(5)
```

```
X = data['review']
y=(data['sentiment'] == 'positive')
```

```
y=y.astype(int)
```

```
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download("stopwords")
nltk.download('punkt')
```

```
X[0]
```

```
stop_words = set(stopwords.words('english'))
```

```
cleaned_token = []
for l in X:
  token = word_tokenize(l)
  token = [ t.lower() for t in token if t.isalpha() and t.lower() not in stop_words]
  cleaned_token.append(token)
```

```
len(cleaned_token)
```

```
import gensim
```

```
model = gensim.models.Word2Vec(sentences=cleaned_token)
```

```
word_to_index={w:i for i, w in enumerate(model.wv.index2word)}
```

```
word_to_index['science']
```

```
model.wv.vectors[945]
```

```
model.wv['science']
```

```
num_words = len(model.wv.vocab)
```

```
index_token=[]
for c in cleaned_token:
  l=[]
  for t in c:
    l.append(word_to_index.get(t, 0))
  index_token.append(l)
```

```
max_l = max([len(s.split()) for s in X])
```

```
max_l
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(index_token,y, test_size=0.2)
```

```
from keras.preprocessing.sequence import pad_sequences
```

```
X_train_pad = pad_sequences(X_train, maxlen=max_l, padding='post')
X_test_pad = pad_sequences(X_test, maxlen=max_l, padding='post')
```

```
from keras.models import Sequential
from keras.layers import Dense, Embedding, GRU
from keras.initializers import Constant
embedding_dim = 100
```

```
smodel = Sequential()
smodel.add(Embedding(num_words, embedding_dim, input_length=max_l, embeddings_initializer=Constant(model.wv.vectors), trainable=False))
smodel.add(GRU(units=32, dropout=0.2, recurrent_dropout=0.2))
smodel.add(Dense(1, activation='sigmoid'))
smodel.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
smodel.summary()
```

```
smodel.fit(X_train_pad, y_train, batch_size=128, epochs=25, verbose=2)
```
