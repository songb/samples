```
import numpy as np
import pandas as pd
import nltk
nltk.download("stopwords")
```

```
from nltk.corpus import stopwords

stop_wrods = stopwords.words("english")
```

```
sample_text_1="bitty bought a bit of butter"
sample_text_2="but the bit of butter was a bit bitter"
sample_text_3="so she bought some better butter to make the bitter butter better"

corp=[sample_text_1,sample_text_2,sample_text_3]
no_docs=len(corp)
```

```
from keras.preprocessing.text import one_hot
vocab_size=50 
encod_corp=[]
for i,doc in enumerate(corp):
    encod_corp.append(one_hot(doc,vocab_size))
    print("The encoding for document",i+1," is : ",one_hot(doc,vocab_size))
```

```
nltk.download("punkt")

```

```
ml=0
for doc in encod_corp:
  ml = max(ml, len(doc))
```

```
from keras.preprocessing.sequence import pad_sequences
```

```
pad_corp = pad_sequences(encod_corp, maxlen=ml, padding="post")
```

```
from keras.layers import Dense, Flatten, Embedding, Input
from keras.models import Sequential, Model
```

```
input = Input(shape=(ml,), dtype="float")
word_embed = Embedding(input_dim=vocab_size, output_dim=8, input_length=ml)(input)
word_flat=Flatten()(word_embed)
emb_model = Model([input], word_flat)
emb_model.summary()
```

```
emb_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```
embeddings = emb_model.predict(pad_corp)
```

```
embeddings = embeddings.reshape(-1, ml, 8)
```
